```python
import pandas as pd
import os

# 파일 경로 설정
file_paths = [
    os.path.join(os.getcwd(), "영덕동_강수_202311_202409.csv"),
    os.path.join(os.getcwd(), "영덕동_강수형태_202311_202409.csv"),
    os.path.join(os.getcwd(), "영덕동_기온_202311_202409.csv"),
    os.path.join(os.getcwd(), "영덕동_습도_202311_202409.csv"),
    os.path.join(os.getcwd(), "영덕동_풍속_202311_202409.csv"),
    os.path.join(os.getcwd(), "영덕동_풍향_202311_202409.csv")
]

# CSV 파일을 읽고 데이터프레임 생성
data_frames = []
for i, file_path in enumerate(file_paths):
    df = pd.read_csv(file_path).dropna(axis=0)
    df = df.dropna(axis=0)
    df.columns = ['일', '시간', f'value_{i}']  # 열 이름 변경
    df['시간'] = df['시간'].astype(int)
    df.set_index(['일', '시간'], inplace=True)  # 인덱스를 'day', 'hour'로 설정
    data_frames.append(df)

# concat으로 모든 데이터프레임 병합
weather_df= pd.concat(data_frames, axis=1, join='outer').reset_index()  # 'day'와 'hour' 기준으로 병합

weather_df.columns = ['일', '시간', '강수', '강수형태', '기온', '습도', '풍속', '풍향']
# 첫 번째 instance인 merged_data1 생성 이후에 month 열 추가

# day 열이 정수형인지 확인하고 변환
weather_df['일'] = weather_df['일'].astype(int)

# month 열을 초기화
weather_df['월'] = 0

# 설정할 month 값의 순서 정의
month_values = [11, 12] + list(range(1, 10))

# 'day == 1'인 경우에 month 값을 설정하는 작업
month_index = 0
for idx in weather_df[(weather_df['일'] == 1) & (weather_df['시간'] == 0)].index:
    weather_df.loc[idx:, '월'] = month_values[month_index]
    month_index = (month_index + 1) % len(month_values)

weather_df['년'] = 2024

weather_df.loc[(weather_df['월'] == 11) | (weather_df['월'] == 12), '년'] = 2023

# Create the '날짜' column by combining 'year', 'month', and 'day'
weather_df['날짜'] = (
    weather_df['년'].astype(str).str.zfill(4) +  # Ensure year is 4 digits
    weather_df['월'].astype(str).str.zfill(2) +  # Ensure month is 2 digits
    weather_df['일'].astype(str).str.zfill(2)  # Ensure day is 2 digits
)

weather_df.drop(axis=1,columns=['일','월','년'],inplace=True)
# Convert the '날짜' column to datetime for comparison

weather_df['날짜'] = pd.to_datetime(weather_df['날짜'], format='%Y%m%d')

# Create a mask for dates after '2024-09-30'
mask = weather_df['날짜'] <= '2024-09-30'

# Filter the DataFrame to keep only the desired rows
weather_df = weather_df[mask]

# Reset index if needed
weather_df.reset_index(drop=True, inplace=True)

# Display the updated DataFrame
weather_df['시간'] = weather_df['시간'].astype(str)
weather_df
```
