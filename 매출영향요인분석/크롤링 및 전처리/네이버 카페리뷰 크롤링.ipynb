```python
import os
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
from PIL import Image
import io
import time
import numpy as np
import cv2
import os
from tqdm import tqdm
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```

```python
def page_down(num):
    body = browser.find_element(By.CSS_SELECTOR, 'body')
    for i in range(num):
        body.send_keys(Keys.PAGE_DOWN)
def page_down_detail():
    element = browser.find_elements(By.CSS_SELECTOR, 'body')
    element=element[1]
    element.send_keys(Keys.END)  
def search_iframe():
    browser.switch_to.default_content()
    browser.switch_to.frame("searchIframe")

def entry_iframe():
    browser.switch_to.default_content()
    WebDriverWait(browser, 10).until(EC.frame_to_be_available_and_switch_to_it((By.ID, "entryIframe")))
    time.sleep(1)
```

```python
how=[]
wating_time=[]
purpose=[]
who=[]
keward_text=[]
store=[]
```

```python
options = Options()
options.add_argument('--headless')
browser = webdriver.Chrome()
browser.get('https://map.naver.com/p?c=15.00,0,0,0,dh')
time.sleep(2)
incheon_input=browser.find_elements(By.CSS_SELECTOR,'input[type=text]')
time.sleep(1)
incheon_input[1].send_keys('인천광역시 중구 카페')
incheon_input[1].send_keys(Keys.ENTER)

# 카페 이름 수집
time.sleep(2)
out = []

for i in range(1, 10):
    search_iframe()
    time.sleep(2)
    name_link = WebDriverWait(browser, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 
                                                                    f'#_pcmap_list_scroll_container > ul > li:nth-child({i}) > div.CHC5F > a.tzwk0 > div > div > span.place_bluelink.TYaxT')))
    name_link.click()
    time.sleep(1)
    entry_iframe()

    # "리뷰 더보기" 버튼을 찾고 클릭
    element2 = browser.find_element(By.XPATH, '//span[text()="리뷰"]')
    element2.click()
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight/4);")

    for k in range(10):
            time.sleep(1)
            while True:
                browser.execute_script("window.scrollTo(0, document.body.scrollHeight/4);")
                try:
                    # 요소가 보일 때까지 기다림 (타임아웃은 10초로 설정)
                    element3 = WebDriverWait(browser, 10).until(
                        EC.visibility_of_element_located((By.XPATH, '//a[@class="fvwqf"]')))
                    # 요소가 보이면 반복문 종료
                    break
                except:
                    # 요소가 아직 보이지 않으면 계속 스크롤
                    pass
            review = WebDriverWait(browser, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR,
                                                                                '#app-root > div > div > div > div:nth-child(6) > div:nth-child(3) > div.place_section.k1QQ5 > div.place_section_content > ul')))
            review_text = review.find_elements(By.TAG_NAME, 'li')
            for text_ind in review_text:
                review_text_ind = text_ind.text
                print(review_text_ind)
                out.append(review_text_ind)
            page_source = browser.page_source
            bs = BeautifulSoup(page_source, 'html.parser')
            allDiv=bs.find_all('li',class_='owAeM')
            for review_box in allDiv:
                store.append(f'카페 {i}')
                try:
                    all_review_box=review_box.find_all('div',class_='MnhVd')
                    if review_box:
                        for review_box2 in all_review_box:
                            listing=[]
                        if review_box2:
                            for review_box_method in review_box2:
                                rr=review_box_method.find_all('span',class_='xalSr')
                                for r in rr:
                                    listing.append(r.text)
                                try:
                                    how.append(listing[0])
                                except:
                                    how.append('평가 없음')
                                try:
                                    wating_time.append(listing[1])
                                except:
                                    wating_time.append('평가 없음')
                                try:
                                    purpose.append(listing[2])
                                except:
                                    purpose.append('평가 없음')
                                try:
                                    who.append(listing[3])
                                except:
                                    who.append('평가 없음')
                except:
                    how.append('평가 없음')
                    wating_time.append('평가 없음')
                    purpose.append('평가 없음')
                    who.append('평가 없음')
                try:
                    keward_box = review_box.find_all('div',class_='ERkm0')
                    for k in keward_box:
                        try:
                            keward = k.find('span')
                            keward_text.append(keward.text)
                        except:
                            keward_text.append('평가 없음')
                except:
                    keward_text.append('평가 없음')
            element3.click()
```

```python
cafe_df = pd.DataFrame(data=[], columns=['가게','방문목적','예약여부','대기시간','동행자','대표키워드'])
# k개의 리뷰를 수집합니다.
for i in range(len(keward_text)):
    tmp = []
    tmp.append(store[i])
    tmp.append(purpose[i])
    tmp.append(how[i])
    tmp.append(wating_time[i])
    tmp.append(who[i])
    tmp.append(keward_text[i])
    tmp = pd.DataFrame(data=[tmp], columns = cafe_df.columns)
    cafe_df = pd.concat([cafe_df,tmp])
    

cafe_df.reset_index(inplace=True, drop=True)
cafe_df
```
